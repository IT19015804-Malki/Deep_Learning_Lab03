```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
im = cv2.imread('/pictures/sample_data/lenna.png')
```

```python
im.shape
```

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

```python
v_kernel = np.array([
    [-5, 0, 5],
    [-5, 0, 5],
    [-5, 0, 5]
])
```

```python
h_kernel = np.array([
    [-2, -1, -2],
    [2, 0, 2],
    [2, 1, 2]
])
```

```python
v_image = cv2.filter2D(im, -5, v_kernel)
h_image = cv2.filter2D(im, -5, h_kernel)
```

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```
